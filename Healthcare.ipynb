```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census = pd.read_csv("census_2011.csv")
census.head()
```

```python
print(census.info())
```

```python
census.describe()
```

```python
census.isnull().sum()
```

```python
census.ffill(inplace = True)
```

```python
census
```

```python
census.isnull().sum()
```

```python
census.plot()
```

```python
state_counts = census['State name'].value_counts()
plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar')
plt.title("Random")
plt.show()
```

```python
state_population = census.groupby('State name')['Population'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(state_population['State name'], state_population['Population'])
plt.xticks(rotation = 90)
plt.show()
```

```python
state_gender = census.groupby('State name')[['Male', 'Female']].sum().reset_index()
plt.figure(figsize= (10, 6))
plt.bar(state_gender['State name'], state_gender['Male'], label="Male")
plt.bar(state_gender['State name'], state_gender['Female'], bottom=state_gender['Male'], label = "Female")
plt.title('Male and Female population per state')
plt.xlabel("State")
plt.ylabel("Population")
plt.xticks(rotation = 90)
plt.show()
```

```python
import numpy as np
state_gender = census.groupby('State name')[['Literate', 'Male_Literate', 'Female_Literate']].sum().reset_index()

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(20, 10))

# Set the x-axis values (states)
x = np.arange(len(state_gender['State name']))

# Set the bar width
bar_width = 0.35

# Create the male and female bars
literate_bars = ax.bar(x - bar_width/2, state_gender['Literate'], bar_width, label='Literate')
male_literate = ax.bar(x, state_gender['Male_Literate'], bar_width, label = 'Male Literate')
female_bars = ax.bar(x + bar_width/2, state_gender['Female_Literate'], bar_width, label='Female')

# Set the x-axis tick labels
ax.set_xticks(x)
ax.set_xticklabels(state_gender['State name'], rotation=90)

# Set the title and labels
ax.set_title('Male and Female Population per State')
ax.set_xlabel('State')
ax.set_ylabel('Population')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
